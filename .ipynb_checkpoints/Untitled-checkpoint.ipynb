```python
import pandas as pd
df=pd.read_csv('owid-covid-data.csv')
df=df[df['new_cases'].notna()]
```

```python
keys=df['date'].unique()
vals=df.groupby('date')['new_cases'].sum()
keys=keys[200:]
vals=vals[200:]
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(15, 4))
# markersize=3, markeredgecolor='black'
plt.scatter(keys, vals, marker='o', s=5)
plt.xticks([])
plt.show()
```
